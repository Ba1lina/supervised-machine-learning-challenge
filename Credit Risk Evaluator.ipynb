## Split the Data into Training and Testing Sets

```python
# Classify x & y
for col in lending_data.columns:
    if lending_data[col].dtype == 'object':
        lending_data[col] = pd.to_numeric(lending_data[col], errors='coerce')
        
X = lending_data.drop('loan_status', axis=1)
y = lending_data['loan_status'] != 1

# Split the data into X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Create, Fit and Compare Models

Create a Logistic Regression model, fit it to the data, and print the model's score. Do the same for a Random Forest Classifier. You may choose any starting hyperparameters you like. 

Which model performed better? How does that compare to your prediction? Write down your results and thoughts in the designated markdown cell.

```python
# Train a Logistic Regression model and print the model score
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
print(f"Training Score: {classifier.score(X_train, y_train)}")
print(f"Testing Score: {classifier.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

*After training both models it is evident that the Logistic Regression model performed better which is contrary to my prediction. Not only is the test score higher but the difference between Logistic Regression training and testing scores is smaller.*
